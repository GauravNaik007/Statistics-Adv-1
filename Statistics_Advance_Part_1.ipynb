# **Theory**

1.  What is a random variable in probability theory?
  - A random variable is a variable whose values are determined by the outcomes of a random experiment.

  - It assigns numerical values to the outcomes.

  - Two main types: discrete and continuous.

  - Example: Rolling a die → outcome (1 to 6) = discrete random variable.
---
2. What are the types of random variables?
  - Discrete Random Variable

    - Takes countable values.

    - Example: Number of heads in 3 coin tosses.

  - Continuous Random Variable

    - Takes any value within a range or interval.

    - Example: Height of students in a class.
---
3. What is the difference between discrete and continuous distributions?
  - Discrete Distribution

    - Deals with countable outcomes.

    - Probability is assigned to specific values.

  - Continuous Distribution

    - Deals with measurable outcomes.

    - Probability is calculated over intervals (area under the curve).
---
4. What are probability distribution functions (PDF)?
  - A PDF describes the likelihood of a random variable taking on a particular value.

  - For discrete variables: Called PMF (Probability Mass Function).

  - For continuous variables: PDF gives probability density (not actual probability).

  - The area under the PDF curve gives the probability over an interval.
---
5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - PDF

    - Shows the density or probability at a specific value (only for continuous).

    - Used to visualize the shape of the distribution.

  - CDF

    - Shows the probability that a variable is less than or equal to a value.

    - Always increases and ranges from 0 to 1.
---
6. What is a discrete uniform distribution?
  - All outcomes have equal probability.

    - Example: Rolling a fair die – each number (1–6) has a 1/6 chance.

  - Used when all events are equally likely.
---
7. What are the key properties of a Bernoulli distribution?
  - Only two outcomes: success (1) or failure (0).

  - Probability of success = p; probability of failure = 1 – p.

  - Mean = p; Variance = p(1 – p).

  - Example: Tossing a coin (Head = 1, Tail = 0).
---
8. What is the binomial distribution, and how is it used in probability?
  - Used for multiple independent Bernoulli trials.

  - Counts the number of successes in fixed trials.

  - Parameters: n (number of trials), p (probability of success).
---
9. What is the Poisson distribution and where is it applied?
  - Models the number of events in a fixed interval (time, space).

  - Events must be independent.

  - Used for rare events like:

    - Number of calls at a call center per hour.

    - Number of accidents in a day.
---
10. What is a continuous uniform distribution?
  - All values in a range have equal probability.

  - Example: Choosing a random number between 2 and 10.

  - Probability is evenly spread out.
---
11. What are the characteristics of a normal distribution?
  - Symmetrical bell-shaped curve.

  - Mean = Median = Mode.

  - Follows empirical rule: 68%-95%-99.7% rule.

  - Example: Heights of people, exam scores.
---
12. What is the standard normal distribution, and why is it important?
  - A normal distribution with mean = 0 and standard deviation = 1.

  - Used to compare different normal distributions.

  - Helps in calculating Z-scores.
---
13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - CLT says: the sampling distribution of the sample mean becomes normal as sample size increases (n ≥ 30).

  - Useful even if population is not normal.

  - Important for hypothesis testing and confidence intervals.
---
14. How does the Central Limit Theorem relate to the normal distribution?
  - we can use the normal distribution to approximate sampling distributions.

  - Helps simplify analysis of real-world data.
---
15. What is the application of Z statistics in hypothesis testing?
  - Used to test hypotheses about population means when standard deviation is known.

  - Compares sample mean to population mean.

  - Helps calculate p-values.
---
16. How do you calculate a Z-score, and what does it represent?
  - Formula: Z = (X – μ) / σ

  - X = sample value, μ = mean, σ = standard deviation.

  - Shows how many standard deviations a value is from the mean.

  - Z = 0 → exactly at the mean.
---
17. What are point estimates and interval estimates in statistics?
  - Point Estimate – A single value (e.g., sample mean) as an estimate.

  - Interval Estimate – A range of values (e.g., 95% confidence interval).

  - Interval gives more reliability.
---
18. What is the significance of confidence intervals in statistical analysis?
  - Shows the range in which the true population parameter lies.

  - Helps express uncertainty.

  - Example: "We are 95% confident that the true mean lies between 70 and 80.
---
19. What is the relationship between a Z-score and a confidence interval?
  - Z-scores are used to calculate confidence intervals.

  - Higher confidence → wider interval → larger Z-score.

  - Example: 95% CI uses Z = 1.96.
---
20. How are Z-scores used to compare different distributions?
  - Converts different data to a common scale (standard normal).

  - Helps compare values from different units or populations.

  - Example: Compare student scores from two different schools.
---
21. What are the assumptions for applying the Central Limit Theorem?
  - Samples must be independent.

  - Sample size should be large (usually n ≥ 30).

  - Population should have finite mean and variance.
---
22. What is the concept of expected value in a probability distribution?
  - It is the average value a random variable takes in the long run.

  - Formula (discrete): E(X) = Σ [x * P(x)]

  - Example: Expected value of a fair die = (1+2+3+4+5+6)/6 = 3.5
---
23. How does a probability distribution relate to the expected outcome of a random variable?
  -  distribution shows all possible outcomes and their probabilities.

  - Expected outcome is the weighted average (expected value) of all outcomes.

  - It gives a prediction of the "long-run" average result.
